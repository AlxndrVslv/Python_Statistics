```python
import pandas as pd
```

## Задание 1. Загрузка данных
Изучить представленный набор данных на основе описания его столбцов в файле “horse_data.names” , загрузить его и оставить 8 столбцов для дальнейшего изучения: surgery?, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.

```python
df = pd.read_csv('horse_data.csv', header = None)
df.columns = [
    'surgery?',
    'Age',
    'Hospital Number',
    'rectal temperature',
    'pulse',
    'respiratory rate',
    'temperature of extremities',
    'peripheral pulse',
    'mucous membranes',
    'capillary refill time',
    'pain',
    'peristalsis',
    'abdominal distension',
    'nasogastric tube',
    'nasogastric reflux',
    'nasogastric reflux PH',
    'rectal examination',
    'abdomen',
    'packed cell volume',
    'total protein',
    'abdominocentesis appearance',
    'abdomcentesis total protein',
    'outcome',
    'surgical lesion?',
    'type of lesion 1',
    'type of lesion 2',
    'type of lesion 3',
    'cp_data'
]
df.head()
```

Обработаем дата-фрейм:
<tr>
<li>- оставим нужные колонки; </li>
<li>- заменим знаки вопроса на цифровые значения, которые больше не встречаются; </li>
<li>- изменим тип данных на числовой для возможности проведения расчетов. </li>
</tr>

```python
df_estimated = df[['surgery?', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome']].replace('?', None)
df_estimated['surgery?'] = pd.to_numeric(df_estimated['surgery?'])
df_estimated['rectal temperature'] = pd.to_numeric(df_estimated['rectal temperature'])
df_estimated['pulse'] = pd.to_numeric(df_estimated['pulse'])
df_estimated['respiratory rate'] = pd.to_numeric(df_estimated['respiratory rate'])
df_estimated['temperature of extremities'] = pd.to_numeric(df_estimated['temperature of extremities'])
df_estimated['pain'] = pd.to_numeric(df_estimated['pain'])
df_estimated['outcome'] = pd.to_numeric(df_estimated['outcome'])
df_estimated.head()
```

## Задание 2. Первичное изучение данных
Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
df_estimated.info()
```

По категориальным атрибутам найдем значения, которые они принимают:

```python
print(df_estimated['surgery?'].value_counts())
print(df_estimated['Age'].value_counts())
print(df_estimated['temperature of extremities'].value_counts())
print(df_estimated['pain'].value_counts())
print(df_estimated['outcome'].value_counts())
```

1. Видим, что признак Age принимает значение 9, которого нет в описании столбцов. 
По суте 9 является выбросом для атрибута Age:

```python
df_outliers_age = df_estimated[df_estimated['Age'] == 9]
df_outliers_age.head()
```

2. Выбросов для атрибутов surgery?, temperature of extremities, pain, outcome нет, т.к. они принимают значения строго своих категорий

3. Выбросы для атрибута rectal temperature:

```python
q1 = df_estimated['rectal temperature'].quantile(0.25)
q3 = df_estimated['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df_outliers_rt = df_estimated[(~df_estimated['rectal temperature'].between(lower_bound, upper_bound)) & (~df_estimated['rectal temperature'].isna())]
df_outliers_rt
```

4. Выбросы для атрибута pulse:

```python
q1 = df_estimated['pulse'].quantile(0.25)
q3 = df_estimated['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df_outliers_pulse = df_estimated[(~df_estimated['pulse'].between(lower_bound, upper_bound)) & (~df_estimated['pulse'].isna())]
df_outliers_pulse
```

5. Выбросы для атрибута respiratory rate:

```python
q1 = df_estimated['respiratory rate'].quantile(0.25)
q3 = df_estimated['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df_outliers_rr = df_estimated[(~df_estimated['respiratory rate'].between(lower_bound, upper_bound)) & (~df_estimated['respiratory rate'].isna())]
df_outliers_rr
```

## Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
df_filled = df_estimated.copy()
```

```python
df_filled.info()
```

1. Для атрибутов surgery?, outcome отсутствуют по одному значению. Следовательно удаление по одной записи для каждого пустого атрибута не должно привести к искажениям результата.

```python
df_filled.dropna(subset = ['surgery?', 'outcome'], inplace = True)
df_filled.info()
```

2. Атрибуты temperature of extremities, pain являются категориальными. Поэтому пропущенные значения по ним логичнее всего заполнить наиболее часто встечающимися значениями, т.е. модой.

```python
df_filled['temperature of extremities'].fillna(df_filled['temperature of extremities'].mode()[0], inplace = True)
df_filled['pain'].fillna(df_filled['pain'].mode()[0], inplace = True)
df_filled.info()
```

3. Для атрибутов rectal temperature, pulse, respiratory rate проверим как сильно изменятся средние статистики при замене пустых значений:<br/>
   а) средним;<br/>
   б) медианой;<br/>
   в) модой

```python
df_filled_mean = df_filled.copy()
df_filled_mean['rectal temperature'].fillna(df_filled_mean['rectal temperature'].mean(), inplace = True)
df_filled_median = df_filled.copy()
df_filled_median['rectal temperature'].fillna(df_filled_median['rectal temperature'].median(), inplace = True)
df_filled_mode = df_filled.copy()
df_filled_mode['rectal temperature'].fillna(df_filled_mode['rectal temperature'].mode()[0], inplace = True)
print('При замене средним:')
print(f"Медиана изменилась на {round(abs(df_filled['rectal temperature'].median() - df_filled_mean['rectal temperature'].median()), 2)}")
print(f"Мода изменилась на {round(abs(df_filled['rectal temperature'].mode()[0] - df_filled_mean['rectal temperature'].mode()[0]), 2)}\n")
print('При замене медианой:')
print(f"Среднее изменилось на {round(abs(df_filled['rectal temperature'].mean() - df_filled_median['rectal temperature'].mean()), 2)}")
print(f"Мода изменилась на {round(abs(df_filled['rectal temperature'].mode()[0] - df_filled_median['rectal temperature'].mode()[0]), 2)}\n")
print('При замене модой:')
print(f"Среднее изменилось на {round(abs(df_filled['rectal temperature'].mean() - df_filled_mode['rectal temperature'].mean()), 2)}")
print(f"Медиана изменилась на {round(abs(df_filled['rectal temperature'].median() - df_filled_mode['rectal temperature'].median()), 2)}\n")
```

Для атрибута rectal temperature:
<li>- при замене средним общее изменение 0,2; </li>
<li>- при замене медианой общее изменение 0,21; </li>
<li>- при замене модой общее изменение 0,23; </li>
Соответственно, в данном атрибуте пустые значения будем заменять средним:

```python
df_filled['rectal temperature'].fillna(df_filled['rectal temperature'].mean(), inplace = True)
df_filled.info()
```

```python
df_filled_mean = df_filled.copy()
df_filled_mean['pulse'].fillna(df_filled_mean['pulse'].mean(), inplace = True)
df_filled_median = df_filled.copy()
df_filled_median['pulse'].fillna(df_filled_median['pulse'].median(), inplace = True)
df_filled_mode = df_filled.copy()
df_filled_mode['pulse'].fillna(df_filled_mode['pulse'].mode()[0], inplace = True)
print('При замене средним:')
print(f"Медиана изменилась на {round(abs(df_filled['pulse'].median() - df_filled_mean['pulse'].median()), 2)}")
print(f"Мода изменилась на {round(abs(df_filled['pulse'].mode()[0] - df_filled_mean['pulse'].mode()[0]), 2)}\n")
print('При замене медианой:')
print(f"Среднее изменилось на {round(abs(df_filled['pulse'].mean() - df_filled_median['pulse'].mean()), 2)}")
print(f"Мода изменилась на {round(abs(df_filled['pulse'].mode()[0] - df_filled_median['pulse'].mode()[0]), 2)}\n")
print('При замене модой:')
print(f"Среднее изменилось на {round(abs(df_filled['pulse'].mean() - df_filled_mode['pulse'].mean()), 2)}")
print(f"Медиана изменилась на {round(abs(df_filled['pulse'].median() - df_filled_mode['pulse'].median()), 2)}\n")
```

Для атрибута pulse:
<li>- при замене средним общее изменение 28; </li>
<li>- при замене медианой общее изменение 16,64; </li>
